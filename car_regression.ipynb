```python
from sklearn.impute import SimpleImputer
import numpy as np

SI = SimpleImputer(strategy='mean', missing_values=np.nan)
data['normalized-losses'] = SI.fit_transform(data[['normalized-losses']])
data['horsepower'] = SI.fit_transform(data[['horsepower']])
data['price'] = SI.fit_transform(data[['price']])
```

```python
X=data
y=df.price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
```

```python
# For train data
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
tr_Xtrain=ss.fit_transform(X_train)
```

```python
# For test data
tr_Xtest=ss.transform(X_test)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
```

```python
dtr=DecisionTreeRegressor()
dtr.fit(X_train,y_train)
print(dtr.score(X_train,y_train))
print(dtr.score(X_test, y_test))
```

```python
sgr=SGDRegressor()
sgr.fit(X_train,y_train)
print(sgr.score(X_train,y_train))
print(sgr.score(X_test, y_test))
```

```python
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
print(rfr.score(X_train, y_train))
print(rfr.score(X_test, y_test))
```

```python
gbr=GradientBoostingRegressor()
gbr.fit(X_train, y_train)
print(gbr.score(X_train, y_train))
print(gbr.score(X_test, y_test))
```
